```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import zipfile
import io
zip_ref = zipfile.ZipFile('API_SP.POP.TOTL_DS2_en_csv_v2_23.zip', 'r')
zip_ref.extractall('data')  # Extracts the contents into the 'data' folder
zip_ref.close()
```

```python
import pandas as pd

df = pd.read_csv('data/API_SP.POP.TOTL_DS2_en_csv_v2_23.csv', skiprows=4)
print(df.head())
```

```python

plt.figure(figsize=(10, 6))
plt.hist(df['2019'], bins=20, edgecolor='black')
plt.title('Distribution of Population in 2019')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
